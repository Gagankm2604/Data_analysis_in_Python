```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_excel('Players.xlsx')
print(df.head())

df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()

df.dropna(inplace=True)
```

```python
df.columns
```

```python
#analysing and plotting the data of right and left handed play

df['Batting_Hand'].value_counts().plot(kind='pie',color=['red','blue'])

```

```python
# Bowling skill analysis and plotting

df['Bowling_Skill'].value_counts().plot(kind='pie',colors=['green','yellow','orange','purple','pink'])
```

```python
df['Country'].value_counts().plot(kind='bar',color='green',edgecolor='black')
plt.title('Country wise players distribution')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.show()
```
